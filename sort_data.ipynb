## Sorting data

```python
import pandas as pd

# set the row index by name
data = pd.read_csv('Data/sales.csv', index_col='State')
```

```python
# sort default
data.sort_values(by='Sales')
```

```python
# sort multiple column
data.sort_values(by=['Sales', 'Profit'], ascending=False)
```

```python
data['Market'].sort_index(ascending=False)
```
